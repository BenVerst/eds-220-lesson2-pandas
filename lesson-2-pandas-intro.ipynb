# Series

# Creating a pandas Series

Several ways of creating a pandas Series. 
For now, we will create series using:

```
s = pd.Series(data, index=index)
```

- `data` = numpy array (or a list of objects that can be converted to NumPy types)
- `index` = a list of indices of same lenght as data


````python
# EX: a pandas series from a numpy array

# np.arange() function constructs an array of consecutive integers
np.arange(3)
````

````python
# we can use this to create a pandas Series
pd.Series(np.arange(3), index=['a','b','c'])
````

What kind of parameter is `index`? 

A: an optional parameter, there is a default value to it. 

If we don't specify `index`, the default is to start the index from 0.

Example:

````python
# create a series from a list of strings with default index
pd.Series(['EDS 220', 'EDS 222', 'EDS 223', 'EDS 242'])
````

# Dataframes

`pandas.DataFrame`:

- most used object in `pandas`
- represents tabular data (think of a spreadsheet)
- each column is a `pandas.Series`

# Createing a `pandas.DataFrame`

*Many ways of creating a dataframe*. Let's see one. 

Remember dictionaries? They are sets of key-value pairs:
```
{ key1 : value1,
  key2 : value2
}
```

Think of a `pd.DataFrame` as a dictionary where:

- keys = column names
- values = column values

We can create a dataframe like this:

````python
# initialize dictionary with columns' data
d = {'col_name_1' : np.arange(3),
     'col_name_2' : [3.1, 3.2, 3.3]
    }
d
````

````python
# create data frame
df = pd.DataFrame(d)
df
````

# In-place operations
Let's rename the data frame's columns.
We can use a dataframe *mehtod* called `rename`. 
`rename` takes in as an input a dictionary:

```
{ 'col_1_old_name' : 'col_1_new_name',
  'col_2_old_name' : 'col_2_new_name'
}
```

````python
# define new column names
col_names = { 'col_name_1' : 'col1',
              'col_name_2' : 'col2'
            }
# rename using rename
df.rename(columns = col_names)
````

````python
# take a look at our dataframe
df
````

Nothing changed! 
`df.rename()` doesn't change the column names *in place*, meaning it doesn't modify the object itself. Instead, it created a new object as an output. 

Assign ouput back to dataframe to actually change it:

````python
df = df.rename(columns = col_names)
df
````

## selecting multiple columns using a list of column names
The syntax for selecting multiple columns using a list of column names is:
```
df[['col1', 'col2', 'col3']]
```
The list of column names `['col1', 'col2', 'col3']` goes inside the selection brackets `[]`. That's why there are double brackets.

Notice each column name is a string.

Example: select change in clacial volume in Europe and Asia

````python
# select columns with names "europe" and "asia"
europe_asia = df[['europe', 'asia']]

# print type of resulting selection
print(europe_asia)

# check the shape of the selection
print(europe_asia.shape)

# see our selection
europe_asia
````

## selecting multiple columns using a slice

Use `loc` selection. The general syntax for `loc` is 
```
df.loc[ row-selection, column-selection]
```
where
- `row-selection` = rows we want to subset from `df`
- `column-selection` = columns we want to subset from `df`

Example: select change in glacial volume per year in all regions.

````python
# select all columns betwwen europe and antarcctica
df.loc[ : , "europe" : "antarctica"]
````

Notes about this:
- `row-selection` = `:` colon = means select all the rows
- `column_selection` = slice `'europe:antarctica'`
- we get both ends of the slice in the selected datafram. Different from slicing in base Python and NumPy, where the endpoint is *not* included in the slice.

## Select rows
### Selecting rows using a condition

General syntax:
```
df[ condition_on_rows]
```

Example: we are interested in data after 1996

````python
# select all rows with year > 1996
df[ df['year']>1996 ]
````

Let's break it down:
- in the `df[condition_on_rows]` syntax, we have that `condition_on_rows` = `df['year']>1996`
- `df['year']>1996` checks which rows have values greater than 1996 in the year column.

````python
df['year']>1996
````

- `df['year']>1996` is a `pd.Series` with boolean values (True Flase)
- `pd.Series` with boolean values (True False) are often called **masks**
- when we pass a `pd.Series` with boolean values (mask) to the selection brackets [] we keep only those rows with a `True` value in the pd.Series/mask

Example: select data from years 1970 to 1979 (including 79)

````python
df[ df.year.isin(range(1970, 1980)) ]
````

````python
df.year.isin(range(1970, 1980))
````

## Select rows using multiple conditions

We can combine multiple conditions by surrounding each on in parenthesis `()` and using the **or operator** `|` and the **an operator** `&`

The syntax is:
```
# select rows of df that satisfy condition1 OR condition2
df[ (condition1) | (condition2)]
```
and

```
# select rows of df that satisfy condition1 OR condition2
df[ (condition1) & (condition2)]
```

Example:

````python
# select rows with
# annual sea level rise <0mm OR annual sea level rise >0.8mm

df[ (df["annual_sea_level_rise"]<0) | (df["annual_sea_level_rise"]>0.8) ]
````

## Selecting rows by position

Sometimes we want to select rows on their *actual* position in the data frame.

Use `iloc` selection, syntax is:

```
df.iloc[row-indices]
```

`iloc` = stands for *integer* location based indexing.

Example:

````python
# select the fifth row = index 4

df.iloc[4]
````

````python
# select rows 23 through 30, including 30
df.iloc[23:31]
````

# Selecting rows and columns simultaneously

We can use `loc` (for abels or conditions) or `iloc` (integer positions).

## selecting rows and columns by label or conditions
we use `loc`:

`row-selection` and `column-selection` can be a condition or subset of labels from the index or the column names.

````python
# select change in glacial volume in Europe per year after 2000
df.loc[ df['year']>2000 , ['year', 'europe']]
````
